### Diccionario de datos

los siguientes conceptos son asociados a las columnas de los dataset que utilizaremos en el analisis de datos sobre el acceso a internet en argentina.
 
- **Dial up**: Un tipo de conexión a Internet que utiliza una línea telefónica para establecer una conexión temporal. Es lento en comparación con otras opciones más modernas.

- **Banda ancha fija**: Se refiere a una conexión a Internet de alta velocidad y constante que utiliza cables físicos, como cables coaxiales o de fibra óptica, para transmitir datos.

- **ADSL (Asymmetric Digital Subscriber Line)**: Es una tecnología de banda ancha que utiliza líneas telefónicas convencionales para transmitir datos a velocidades más rápidas que las conexiones de marcado.

- **Cable módem**: Es un dispositivo que permite la conexión a Internet a través de la red de cable de televisión. Proporciona velocidades de conexión más rápidas en comparación con la conexión de marcado.

- **Fibra óptica**: Un tipo de tecnología de conexión a Internet que utiliza cables hechos de fibras ópticas para transmitir datos a través de pulsos de luz. Es extremadamente rápido y proporciona una conexión estable.

- **Wireless (inalámbrico)**: Se refiere a la conexión a Internet sin necesidad de cables físicos. Utiliza tecnologías como Wi-Fi para transmitir datos de manera inalámbrica.

- **Link INDEC**: Hace referencia al acceso a Internet proporcionado por el Instituto Nacional de Estadísticas y Censos (INDEC) en Argentina. Es una conexión a través de un proveedor de servicios de Internet específico.

- **Satélital**: Se refiere a la conexión a Internet que utiliza satélites para transmitir datos. Es una opción viable en áreas donde otras formas de conectividad son limitadas o inexistentes.

- **WiMAX (Worldwide Interoperability for Microwave Access)**: Es una tecnología de conexión a Internet inalámbrica de largo alcance que permite velocidades de banda ancha en áreas extensas.

- **4G**: Es la cuarta generación de tecnología de redes móviles. Proporciona velocidades de conexión más rápidas en comparación con las generaciones anteriores, permitiendo una mejor experiencia de navegación móvil.

- **Telefonía fija**: Hace referencia a la conexión a Internet a través de líneas telefónicas tradicionales. Aunque puede ser más lenta que otras opciones, es ampliamente utilizada en entornos residenciales y comerciales.

En el grafico observamos que a lo largo del tiempo se puede destacar:
 
- Tendencia general: A medida que pasan los años, se observa un aumento en la penetración de Internet, es decir, un mayor número de accesos por cada 100 hogares.

- Crecimiento constante: El crecimiento en la penetración de Internet es constante a lo largo de los años, con aumentos moderados en la mayoría de los casos. En los primeros años, el crecimiento es más gradual, mientras que en los años más recientes, el crecimiento es más pronunciado.

- Aceleración reciente: Se observa una aceleración en el crecimiento de la penetración de Internet en los últimos años, especialmente a partir de 2018. Esto indica un aumento significativo en la adopción de Internet en los hogares.

- Tendencia ascendente: En general, la penetración de Internet muestra una tendencia ascendente a lo largo del período analizado. Cada año, se registra un mayor número de accesos por cada 100 hogares, lo que indica un mayor alcance de Internet en la población.

```python
# Agrupa los datos por provincia y calcula el promedio de accesos por cada 100 hogares
ipp_mean_p = ipp.groupby(['Provincia'])['AccesoX100Hog'].mean().sort_values(ascending=False).reset_index()
ipp_mean_p
```

```python

# Definir los rangos y los colores correspondientes
rango1 = 100  # Porcentaje para el primer rango (100%)
rango2 = 50  # Porcentaje para el segundo rango (50%)
rango3 = 35  # Porcentaje para el tercer rango (35%)
color1 = 'navy'  # Color para el primer rango (azul oscuro)
color2 = 'lightblue'  # Color para el segundo rango (azul claro)
color3 = 'lightgreen'  # Color para el tercer rango (verde claro)
color4 = 'yellow'  # Color para el cuarto rango (amarillo)

# Asignar los colores según los rangos definidos
color_palette = []
for value in ipp_mean_p['AccesoX100Hog']:
    if value >= rango1:
        color_palette.append(color1)
    elif value >= rango2:
        color_palette.append(color2)
    elif value >= rango3:
        color_palette.append(color3)
    else:
        color_palette.append(color4)

# Configurar el gráfico de barras
plt.figure(figsize=(12, 6))
bars = plt.bar(ipp_mean_p['Provincia'], ipp_mean_p['AccesoX100Hog'], color=color_palette)

# Etiquetas y título del gráfico
plt.xlabel('Provincia')
plt.ylabel('Accesos por cada 100 hogares (promedio)')
plt.title('Penetración promedio de Internet por provincia')

# Ajustar la rotación de las etiquetas del eje x 
plt.xticks(rotation=90)

# Agregar los valores en la parte superior de cada barra
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval + 0.5, f'{int(yval)}%', ha='center', va='bottom')

# Mostrar el gráfico
plt.show()
```

En el grafico observamos:

- Variación en la penetración de Internet: Existe una variación significativa en la penetración de Internet entre las provincias. La provincia con la mayor penetración es Capital Federal, con un porcentaje del 113% de accesos por cada 100 hogares, mientras que Formosa registra la menor penetración con un porcentaje del 26% de  accesos por cada 100 hogares.

- Diferencias regionales: Se pueden identificar patrones regionales en la penetración de Internet. la capital federal tienen mas conexiones que hogares con un (113%) y hay 14 provincias por debajo del 50% de penetracion al internet.

- Provincias con alta penetración: Además de Capital Federal, otras provincias como Buenos Aires, Córdoba, Santa Fe y La Pampa también muestran una penetración relativamente alta de Internet. Estas provincias pueden ser consideradas áreas clave para el desarrollo y la adopción de servicios de Internet.

- Provincias con baja penetración: Formosa, Santiago del Estero y Chaco son algunas de las provincias que presentan una menor penetración de Internet. Estas provincias pueden requerir una mayor atención y esfuerzos para mejorar la infraestructura y promover la adopción de servicios de Internet.

Este análisis permite identificar las diferencias en la penetración de Internet entre las provincias y brinda información relevante para comprender las necesidades y oportunidades de cada región en términos de acceso a Internet. Estos datos pueden ser utilizados para desarrollar estrategias de expansión, mejorar la infraestructura y promover la inclusión digital en las provincias con una menor penetración de Internet.

```python

# Extraer las columnas de interés
columnas_velocidad = ['HASTA 512 kbps', '+ 512 Kbps - 1 Mbps', '+ 1 Mbps - 6 Mbps',
                      '+ 6 Mbps - 10 Mbps', '+ 10 Mbps - 20 Mbps', '+ 20 Mbps - 30 Mbps',
                      '+ 30 Mbps', 'OTROS']

# Calcular la suma de accesos para cada categoría de velocidad
suma_accesos = iavp[columnas_velocidad].sum()

# Calcular el total de accesos
total_accesos = suma_accesos.sum()

# Crear el gráfico de barras
plt.figure(figsize=(10, 8))  # Ajustar el tamaño del gráfico
bars = plt.bar(columnas_velocidad, suma_accesos)

# Ajustar la distancia vertical entre los valores encima de las barras
offset = 0.01 * total_accesos
for i, v in enumerate(suma_accesos):
    plt.text(i, v+offset, str(v), ha='center', va='bottom', color='black', fontweight='bold')
    plt.text(i, v-offset*-2, '{:.1f}%'.format(v/total_accesos*100), ha='center', va='bottom', color='red')

# Separar las barras
plt.subplots_adjust(wspace=0.5)

# Personalizar el gráfico
plt.xlabel('Categoría de Velocidad')
plt.ylabel('Cantidad de Accesos')
plt.title('Accesos a Internet por Categoría de Velocidad',pad=30)

# Crear la leyenda con los colores correspondientes
bar_labels = ['Cantidad de Accesos', 'Porcentaje de Accesos']
legend_labels = [plt.Rectangle((0, 0), 1, 1, fc='steelblue', edgecolor='none'), 
                 plt.Rectangle((0, 0), 1, 1, fc='red', edgecolor='none')]
plt.legend(legend_labels, bar_labels, loc='upper right')

# Ajustar el rango del eje y
plt.ylim(0, 1.1 * max(suma_accesos)) 


# Rotar las etiquetas del eje x para mejorar la legibilidad
plt.xticks(rotation=70)

# Mostrar el gráfico
plt.show()
```

- Las velocidades de conexión más bajas, hasta 512 Kbps y de 512 Kbps a 1 Mbps, representan un porcentaje menor del total de usuarios (0.9% y 3.1% respectivamente). Es alentador observar que el número de personas con estas velocidades más lentas está disminuyendo gradualmente.

- Por otro lado, se observa un aumento en el número de usuarios con velocidades más rápidas. Las categorías de 10 Mbps a 30 Mbps representan un porcentaje significativo de los usuarios encuestados. La categoría de 30 Mbps en particular tiene un 23.6% de los usuarios, lo cual indica un incremento en la adopción de velocidades más altas.

- La categoría "OTROS" representa un porcentaje bajo del total de usuarios (1.5%). Esto sugiere que hay una pequeña fracción de usuarios con velocidades de conexión que no se ajustan a los rangos mencionados anteriormente o que no se pueden clasificar fácilmente.

- En resumen, las conclusiones indican una tendencia positiva hacia velocidades de conexión más rápidas, con un aumento en el número de usuarios en los rangos de 10 Mbps a 30 Mbps. Sin embargo, todavía hay una proporción considerable de usuarios con velocidades moderadas, aunque su número está disminuyendo gradualmente.
